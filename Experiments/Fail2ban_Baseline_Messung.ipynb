```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
df_50_1000_254 = pd.read_csv('dns_50_1000_254_correct.csv')
df_50_5000_254 = pd.read_csv('dns_50_5000_254_correct.csv')
df_50_1000_50 = pd.read_csv('dns_50_1000_50_correct.csv')
df_50_5000_50 = pd.read_csv('dns_50_5000_50_correct.csv')
xdp_drop_50_1000_254 = df_50_1000_254[df_50_1000_254['XDP_action  '] == 'XDP_DROP    ']
xdp_pass_50_1000_254 = df_50_1000_254[df_50_1000_254['XDP_action  '] == 'XDP_PASS    ']
xdp_drop_50_5000_254 = df_50_5000_254[df_50_5000_254['XDP_action  '] == 'XDP_DROP    ']
xdp_pass_50_5000_254 = df_50_5000_254[df_50_5000_254['XDP_action  '] == 'XDP_PASS    ']
xdp_drop_50_1000_50 = df_50_1000_50[df_50_1000_50['XDP_action  '] == 'XDP_DROP    ']
xdp_pass_50_1000_50 = df_50_1000_50[df_50_1000_50['XDP_action  '] == 'XDP_PASS    ']
xdp_drop_50_5000_50 = df_50_5000_50[df_50_5000_50['XDP_action  '] == 'XDP_DROP    ']
xdp_pass_50_5000_50 = df_50_5000_50[df_50_5000_50['XDP_action  '] == 'XDP_PASS    ']

fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(5)

plt.plot(xdp_pass_50_1000_254['time     '],xdp_pass_50_1000_254['packets/sec       '],label='pass')
plt.plot(xdp_pass_50_1000_254['time     '],xdp_drop_50_1000_254['packets/sec       '],label='drop')

plt.legend()
plt.xlabel('Sec')
plt.ylabel('PPS')
plt.title('DNS_50_1000_254')
plt.grid(axis='y')
plt.show()
```

```python
def repair_csv(infile,outfile):
    with open(infile,'r') as file:
        new_lines = []
        lines = file.readlines()

        for line in lines:
            new_lines.append(re.sub(r'([0-9]),([0-9])',r'\1\2',line))

        with open(outfile,'w') as out:
            for line in new_lines:
                out.write(line)
    


```
