```python
def cleanTime(df):
    # For some god forsaken reason the time sometimes contains a comma
    # We need to remedy that
    if "," in str(df["time"].iloc[0]):

        df = df.astype({'time': 'string'})
        df["time"] = df.time.str.replace(',', '.')
        df = df.astype({'time': 'float'})

    # Remove time steps outside of the measurement duration
    while (True):
        # Removing all time steps after the measurement ended
        if (df.iloc[-1]['packets/sec'] == 0):
            df = df[:-3]
        else:
            break

    # Check for existence of first empty time step
    if (df.iloc[2]['packets/sec'] == 0):
        while (True):
            # Removing all time steps before the measurement started, except for the first empty one
            # 1 -> Checking for XDP_DROP
            # 2 -> Checking for XDP_PASS
            # 3 -> Checking for XDP_ABORTED
            # 4 -> Checking for XDP_DROP
            # ...
            # Here we use 5, to check for second XDP_PASS with != 0 packets/sec
            # and leave the first time step where each XDP_action packets/sec equals 0
            if (df.iloc[5]['packets/sec'] == 0):
                df = df[3:]
            else:
                break


    return df
```

```python
# Creating baseline measurement graphs

for path, params in paths_simplefail2ban_file.items():
    dfFile = cleanTime(pd.read_csv(path, sep=';'))
    dfShm = cleanTime(pd.read_csv(path.replace("file/file", "shm/shm"), sep=';'))
    dfSock = cleanTime(pd.read_csv(path.replace("file/file", "sock/sock"), sep=';'))

    dfFile = dfFile.rename(columns={'XDP_action': 'XDP_action_file'})
    dfFile_drop = dfFile[dfFile['XDP_action_file'] == 'XDP_DROP']
    dfFile_pass = dfFile[dfFile['XDP_action_file'] == 'XDP_PASS']
    
    dfShm = dfShm.rename(columns={'XDP_action': 'XDP_action_shm'})
    dfShm_drop = dfShm[dfShm['XDP_action_shm'] == 'XDP_DROP']
    dfShm_pass = dfShm[dfShm['XDP_action_shm'] == 'XDP_PASS']
    
    dfSock = dfSock.rename(columns={'XDP_action': 'XDP_action_sock'})
    dfSock_drop = dfSock[dfSock['XDP_action_sock'] == 'XDP_DROP']
    dfSock_pass = dfSock[dfSock['XDP_action_sock'] == 'XDP_PASS']

    fig = plt.figure()
    fig.set_figwidth(20)
    fig.set_figheight(12)
    f_size=15
    rc('text', usetex=True)
    rc('axes', linewidth=2)
    rc('font', weight='bold')
    rc('xtick', labelsize=f_size) 
    rc('ytick', labelsize=f_size)
    
    plt.plot(dfFile_pass['time'][::5], dfFile_drop['packets/sec'][::5],label=r'\textbf{File drop}',linewidth=1.5,color='#6a8532')
    plt.plot(dfFile_pass['time'][::5], dfFile_pass['packets/sec'][::5],label=r'\textbf{File pass}',linewidth=1.5,color='#a1c349')
    
    plt.plot(dfShm_pass['time'][::5], dfShm_drop['packets/sec'][::5],label=r'\textbf{Shm drop}',linewidth=1.5,color='#00193b')
    plt.plot(dfShm_pass['time'][::5], dfShm_pass['packets/sec'][::5],label=r'\textbf{Shm pass}',linewidth=1.5,color='#00377e')
    
    plt.plot(dfSock_pass['time'][::5], dfSock_drop['packets/sec'][::5],label=r'\textbf{Sock drop}',linewidth=1.5,color='#782832')
    plt.plot(dfSock_pass['time'][::5], dfSock_pass['packets/sec'][::5],label=r'\textbf{Sock pass}',linewidth=1.5,color='#cc095d')
    
    title = f'{params.ipStack}: {params.ppsi} invalid traffic from {params.clients} clients. {params.ppsv} valid traffic from 254 clients'
    
    plt.legend(fontsize=f_size, loc='center left', bbox_to_anchor=(1, 00.5))
    plt.xlabel(r'\textbf{Time [Seconds]}',fontsize=f_size)
    plt.ylabel(r'\textbf{Packets per second}',fontsize=f_size)
    plt.title(rf'\textbf{{{title}}}',fontsize=f_size)
    plt.grid(axis='y')

    newPath = path.replace("file/file_", "results/")
    plt.savefig(newPath.replace('.csv','.png'))

    plt.close()
```
